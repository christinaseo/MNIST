```python
def toProbs(i):
    probabilities = np.zeros((10,1))
    probabilities[i] = 1
    return probabilities
```

```python
with gzip.open('mnist.pkl.gz', 'rb') as f:
    train_set, valid_set, test_set = pickle.load(f, encoding='iso-8859-1')

train_x, train_y = train_set

test_x, test_y = test_set

np.reshape(train_y,(50000,1))

train_yf = []

for x in train_y:
    train_yf = np.append(train_yf, toProbs(x))


train_yf = np.reshape(train_yf,(50000,10))

#plt.imshow(train_x[0].reshape((28, 28)), cmap=cm.Greys_r)
#plt.show()
```

```python
def guess(output):
    return np.argmax(output)
```

```python
#create a neural network class
class nn(object):
    def __init__ (self):
        self.inSize = train_x.shape[1]
        self.hiddenSize = 256
        self.outSize = 10
        
        self.ihW = np.zeros(shape = (self.inSize, self.hiddenSize))
        self.hBias = np.zeros(shape=(1, self.hiddenSize))
        self.hoW = np.zeros(shape = (self.hiddenSize, self.outSize))
        self.oBias = np.zeros(shape=(1, self.outSize))
        
        
    def forwardprop(self, inp):
        self.z = np.dot(inp,self.ihW) + self.hBias
        self.z2 = self.sigmoid(self.z)
        self.z3 = np.dot(self.z2,self.hoW) + self.oBias
        final = self.sigmoid(self.z3)
        return final
    
    def sigmoid(self,z):
        return (1 / (1 + np.exp(-z)))
    
    def sigmoid_prime(self,z):
        return z*(1-z)
    
    def backprop(self, diff, output, x, ySize):
        self.dhoW = (1/ySize)*(self.z2.T).dot(diff)
        
        self.dz = np.multiply(((diff.dot(self.hoW.T))), ((self.sigmoid_prime(self.z2))))
        
        self.dihW = (1/ySize)*np.dot(x.T, self.dz)
        
        self.doBias = (1/ySize)*np.sum(diff, axis = 0)
        
        self.dhBias = (1/ySize)*np.sum(self.dz, axis = 0)
        
        self.ihW = self.ihW + self.dihW
        self.hoW = self.hoW + self.dhoW
        self.oBias = self.oBias + self.doBias
        self.hBias = self.hBias + self.dhBias
```
