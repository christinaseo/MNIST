```python
def toProbs(i):
    probabilities = np.zeros((10,1))
    probabilities[i] = 1
    return probabilities
```

```python
#create a neural network class
class nn(object):
    def __init__ (self):
        self.inSize = train_x.shape[1]
        self.outSize = 10
            
        #randomly initialize the weights
        #self.w = np.random.standard_normal((self.inSize, self.outSize))
        #self.b = np.random.standard_normal((1,self.outSize))
        
        self.w = np.zeros(shape=(self.inSize, self.outSize))
        self.b = np.zeros(shape=(1, self.outSize))
        
    def forwardprop(self, inp):
        self.z = np.dot(inp, self.w) +self.b
        #print(np.shape(self.z))
        self.a = self.sigmoid(self.z)
        #print(np.shape(self.a))
        #print(self.a)
        return self.a
    
    def sigmoid(self,z):
        return (1 / (1 + np.exp(-z)))
    
    def sigmoid_prime(self,z):
        return z*(1-z)
    
    def backprop(self, diff, output, x, ySize):
        print("sub" + str(diff))
        self.dw = (1/ySize)*(np.dot(x.T,diff))
        self.db = (1/ySize)*(np.sum(diff)) 
        assert (self.dw.shape == self.w.shape)
        
        print("w before: "+str(self.w))
        
        #PROBLEM IS THAT DW IS WAY TOO SMALL
        print("diff: "+str(np.sum(np.abs(self.w - (self.w - self.dw*lr)))))
        
        self.test = self.w-self.dw*lr
        
        self.w = self.w + self.dw*lr
        self.b = self.b + self.db*lr

        
    def cost(self,yhat,y):
        print("yhat "+ str(yhat))
        yhat = np.reshape(yhat,(50000,1))
        maxi = np.amax(yhat)
        yhat = yhat / maxi
        print(np.shape(yhat))
        print("y "+str(y))
        print(np.shape(y))
        y = np.reshape(y,(50000,1))
        
        y = np.abs(y)
        yhat = np.abs(yhat)
        
        print("positive")
        print("y"+str(y))
        print("yhat"+str(yhat))
        
        print("first part" +str(y*np.log(yhat)))
        print("second "+ str((1-y)*(np.log(1-yhat))))
        temp = (y*np.log(yhat))+((1-y)*np.log(1-yhat))
        print("temp" + str(temp))
        j = (-(1/np.size(y))*np.sum(temp,axis = 1))
        
        print("j"+str(j))
        return j
```
